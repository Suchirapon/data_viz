```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

# Parallel Coordinates 

Coordinates คือ พิกัดที่อยู่บนพื้นที่ X หรือ Y  

Parallel Coordinates คือ การเอา Coordinates มาต่อกัน โดยข้อมูลจะแบ่งเป็น class 

```
pd.plotting.parallel_coordinates(df,'Name') ;
```

## ปรับ parameter : Color

```
pd.plotting.parallel_coordinates(
    df, 'Name', color=('#FB7D62', '#BDF356', '#5695F3') #สี คือ HTML Color จะช่วยให้เราเห็นว่า data เป็นคนละชนิดกัน
)
```

02-rgb-cmyk-1920.jpg

```
pd.plotting.parallel_coordinates(
    df, 'Name', color=('r', 'g', 'b') #ตัวย่อสี R G B เป็น สีของแสง , C M Y K
)
```

## RGB Value

RGB_COLOR.jpg

```
pd.plotting.parallel_coordinates(
    df, 'Name', color=((1,0,0), (0,1,0), (0,0,1)) #rgb value กำหนดปริมาณของแสงที่จะผสมกันได้ ซึ่ง rgb จะต้องมีค่าตั้งแต่ 0-1
)
```

```
pd.plotting.parallel_coordinates(
    df, 'Name', color=((1,0.5,0.8), (0.1,1,0.2), (0.6,0,1)) #rgb value กำหนดปริมาณของแสงที่จะผสมกันได้ ซึ่ง rgb จะต้องมีค่าตั้งแต่ 0-1
)
```

## ปรับ parameter : colormap

- https://matplotlib.org/stable/tutorials/colors/colormaps.html

```
pd.plotting.parallel_coordinates(
    df, 'Name', colormap = 'cool' # เลือกสีจากที่กำหนดไว้แล้ว 
)
```

# เริ่ม Matplotlib

```
from matplotlib import pyplot as plt
```

### แปลง data จาก pandas dataframe เป็น numpy array

```
np_data = df.iloc[:,:-1].to_numpy()
np_data
```

# Visualize array data ด้วย pixel หรือรูปภาพในการโชว์

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15] #กำหนดขนาดของรูปว่าจะให้มีขนาดเท่าใด โดย .rcParams

plt.imshow(np_data[:50,:],interpolation='nearest') # .imshow คือ โชว์ในเป็นรูปภาพ โดยค่ามาก สีเข็ม ค่าน้อย สีอ่อน
# interpolation='nearest' เป็นการแยกให้มองเห็นชัดขึ้น
```

```
plt.imshow(np_data[50:100,:],interpolation='nearest')
```

# การใช้ subplot

```
plt.subplot(1,3,1) # subplot มองว่ามีกี่แถว , มีกี่หลัก , เป็นตัวที่เท่าไหร่
plt.imshow(np_data[:50,:],interpolation='nearest', cmap = 'magma')

plt.subplot(1,3,2) 
plt.imshow(np_data[50:100,:],interpolation='nearest', cmap = 'magma')

plt.subplot(1,3,3) 
plt.imshow(np_data[100:150,:],interpolation='nearest', cmap = 'magma')
```

```
plt.imshow(np_data[:10,:],interpolation='nearest', cmap = 'magma', alpha = 0.5)
# alpha คือ ค่าความโปร่งใส
```
