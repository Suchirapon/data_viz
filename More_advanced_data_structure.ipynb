# HW 5 เขียน function คูณ matrix

แล้ว test กับ matrix ขนาด

- (2,3)*(3,2)

- (4,4)*(4,1)

- (2,2)*(2,2)

```
import numpy as np
```

```
a = np.array([[1,3,5],[2,4,6]]) #matrix 2*3
b = np.array([[1,3],[2,4],[5,6]]) #matrix 3*2
print(a)
print(b)
```

```
c = np.array([[1,3,5,7],[2,4,6,8],[1,2,3,4],[5,6,7,8]]) #matrix 4*4
d = np.array([[5],[4],[3],[2]]) #matrix 4*1
print(c)
print(d)
```

```
e = np.array([[3,5],[2,4]]) #matrix 2*2
f = np.array([[8,6],[9,7]]) #matrix 2*2
print(e)
print(f)
```

```
def matrix(a, b) :
    g = [[0 for row in range(len(b[0]))] for col in range(len(a))]
    for i in range(len(a)) :
        for j in range(len(b[0])) :
            for k in range(len(b)) :
                g[i][j] += a[i][k]*b[k][j]
    mat_ans = numpy.matrix(g)
    return mat_ans
```

```
# matrix (2,3)*(3,2)
print(a)
print(b)
matrix(a,b)
```

```
# matrix (4,4)*(4,1)
print(c)
print(d)
matrix(c,d)
```

```
# matrix (2,2)*(2,2)
print(e)
print(f)
matrix(e,f)
```

## เฉลย HW5

ภาพหน้าจอ (1047).png

```
def mat_mul(A,B) : 
      C = np.zeros((A.shape[0],B.shape[1])) #เตรียมขนาดของ matrix ผลลัพธ์
      for r_a in range(A.shape[0]) :
        for c_b in range(B.shape[1]) :
          for every_element in range(A.shape[1]) :
            C[r_a,c_b] = C[r_a,c_b] + (A[r_a,every_element] * B[every_element,c_b])
      return C
#ออกสอบ
```

```
mat1 = np.array([[1,2,3],[4,5,6]])
mat2 = np.array([[7,8],[9,10],[11,12]])
```

```
mat_mul(mat1,mat2)
```
